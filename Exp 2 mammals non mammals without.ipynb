```python
data=pd.read_csv("C:\\Users\\nisha\\OneDrive\\Documents\\Capture.csv")
data
```

```python
list1=[list(row) for row in data.values]
m=0
nm=0

for i in range(0,len(list1)):
    if(list1[i][5]=="mammals"):
     m+=1
               
    else:
     nm+=1
print("Total number of mammalas",m)
print("Total number of non-mammals",nm)
                
```

```python
def bayesx(X_):
    data1=data[X_]
    prob=[1,1,1]
    list2=[list(row) for row in data1.values]
    for j in range(0,len(X_)-1):
        mam=0
        nmam=0
        for i in range(0,len(list2)):
            if((j==0 and list2[i][j]=="yes") or (j==1 and list2[i][j]=="no") or (j==2 and list2[i][j]=="yes") or (j==3 and list2[i][j]=="no")):
                if(list2[i][4]=="mammals"):
                    mam+=1
                else:
                    nmam+=1
        
        temp=prob.pop(0)
        prob.insert(0,temp*(mam/m))
        temp1=prob.pop(1)
        prob.insert(1,temp*(nmam/nm))
    
    x_prob_mam=prob.pop(0);
    x_prob_nonmam=prob.pop(0);
    total=x_prob_mam + x_prob_nonmam
    prob.insert(0,x_prob_mam/total)
    prob.insert(1,x_prob_nonmam/total)
    print("Mammals",prob[0])
    print("Non Mammals",prob[1])
    
    if(prob[0]>prob[1]):
        print("Mammal")
    else:
        print("Non Mammal")
    
```
